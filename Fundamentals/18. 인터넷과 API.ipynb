#### 1-2. 인터넷
- 컴퓨터들 사이의 TCP/IP(Transmission Control Protocol / Internet Protocol)이라는 통신 프로토콜을 이용해 정보를 주고 받는 컴퓨터 네트워크로 정의된다.

#### 1-3. 소켓 통신
- 엔드포인트 의미
- 소켓과 소켓 사이의 peer-to-peer 연결을 구현하는 프로토콜이 TCP

```python
import socket
with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:     # 소켓 s를 하나 열어서
    s.connect(('www.google.com', 80))          # www.google.com 이라는 호스트의 80번 포트에 접속합니다. 
    cmd = 'GET http://www.google.com HTTP/1.0\r\n\r\n'.encode() 
    s.send(cmd)                 # 접속이 되면 GET xxxxx 형식의 HTTP 프로토콜로 된 명령을 전송합니다. 
    data = s.recv(1024)    # 그리고 1024바이트만큼의 HTTP 응답을 수신합니다. 
print('Received', data)  # 어떤 응답을 수신했는지 볼까요?
```

#### 1-4. 인터넷 프로토콜: TCP/IP
- TCP: Transmission Control Protocol -> 소켓 포트 단위의 송수신 프로토콜
    - Transport 레이어에서 데이터의 흐름과 정확성 여부를 관리한다. 하나의 컴퓨터에 여러개의 소프트웨어가 구동하고 있다면 각각이 다른 포트 번호의 소켓을 가지고 있게 되는데, 이 포트번호를 기준으로 데이터가 타겟 소프트웨어 소켓으로 정확히 전송될 수 있도록 데이터 흐름을 책임지는 것이 TCP의 역할이다.
- IP: Internet Protocol -> 컴퓨터마다 주어지는 IP 주소 단위의 송수신 프로토콜
    - 패킷을 목적지까지 전달한다. 목적지가 되는 컴퓨터(호스트)는 저마다의 IP 주소를 할당받는다. IP는 타겟 주소로 데이터가 정확히 전송되는 것을 책임지는 역할을 하는 프로토콜이다.
- UDP: User Datagram Protocol -> TCP가 두 소켓 간의 쌍방향 독점적 통신이므로 connection 기반의 통신 프로토콜인 반면, UDP는 connection 개념이 없는 비연결형이다. 데이터그램을 일방적으로 전송하기 때문에 전송속도는 TCP보다 빨라서 음성, 영상 데이터 전송에 유리하다.
- Port Number: TCP/IP 기반으로 구동된다는 것은 IP주소와 포트번호를 가지고 통신한다는 뜻과 같은 의미이다.

### 2. API
#### 2-1. Web / HTTP
- HTTP는 요청(Request)하고, 응답(Response)하는 서버/클라이언트 모델
- Request Method (요청 메소드)
    - GET: read, 정보를 요청
    - HEAD: 헤더 정보만 요청
    - POST: create, 정보를 생성, 변형 할때 사용
    - PUT: update, 정보를 수정
    - DELETE: delete, 정보를 삭제
    - OPTIONS : 사용가능한 메소드 확인
- Response Code (응답 코드)
    - 1xx: 조건부 응답
    - 2xx: 성공
    - 3xx: 리다이렉션 완료
    - 4xx: 요청 오류 (클라이언트 문제)
    - 5xx: 서버 오류 (서버 문제)
- Header (헤더)
    - URL: 웹의 자원을 가리키는 위치 지정자
    - URI: 인터넷의 자원을 나타내는 식별자이고 자원을 나타내는 주소
- 인터넷(1960) 하에 FTP, 웹, 이메일 기술 존재
    - 인터넷: 전세계의 거대한 네트워크, 도로!
    - Web(1990): 전세계에 깔린 네트워크를 통해 컴퓨터와 컴퓨터가 데이터를 주고 받을 수 있게 만드는 존재
        - 팀버너스리: 정보를 효율적으로 주고받을 수 있는 시스템 만들어버림! 최초의 웹브라우저, 웹서버 만들었다. 그 사이의 통신규약인 HTTP, 웹언어 HTML도 만들었다.

#### 2-2. API (Application Programing Interface)
- API는 다른 여러 어플리케이션 사이에 간편한 인터페이스를 제공한다.
- 프로그램들이 서로 소통하기 위해 만들어졌다.
- 앱, 서버와 소통할 수 있는 키보드 같은 것들......
- 접근권한이 있는 API도 존재
- 해당 프로그램의 기능을 다른 프로그램이 쓸 수 있게 하는 것이 목적
- 사람은 규칙에 맞게 전송하면 해당 기능을 전송해준다.

#### 2-3. REST API
- Representational State Transfer
- 자원을 URI로 표시하고 해당 자원의 상태를 주고 받는 것을 의미한다.
- REST기반의 규칙들을 지켜서 설계된 API
    - URI는 정보의 자원을 표현해야한다.

```python
# 에러 발생 코드
from urllib.error import URLError, HTTPError, ContentTooShortError

def download(url):
    try:
        html = urllib.request.urlopen(url)
    except (URLError, HTTPError, ContentTooShortError) as e:
        print('Download error', e.reason)
        html = None
    return html

download('https://www.google.com')
```

#### 3-2. 실습

```python
import requests
from bs4 import BeautifulSoup

url = "https://m.stock.naver.com/marketindex/index.nhn"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

country = []
country_1 = soup.findAll('strong')
country_2 = soup.findAll('span', {'class':'stock_item'})
for c in country_1:
    country.append(c.text)
for c in country_2:
    country.append(c.text)
country.remove('컨텐츠 제공업체')
country.remove('증권')
country.remove('국제금')
country.remove('유가WTI')
country.remove('달러인덱스')

price = []
price_0 = soup.findAll('span', {'class': 'stock_price'})
for p in price_0:
    price.append(p.text)
    
import pandas as pd
data = {}
for i in range(len(country)):
    data[country[i]] = price[i]
    
pd.Series(data)
```

### 4. 로그인
#### 4-1. 쿠키와 세션
- 쿠키: 편의를 위함
- 세션: 중요한 정보들, 남발하면 과부화
- 캐시: 가져오는데 비용이 드는 데이터를 한 쪽에 저장

- 쿠키: 쿠키는 HTTP 헤더를 기반으로 이루어 지는 데이터 입니다. "클라이언트 로컬"에 저장되는 Key-Value쌍의 작은 데이터 파일. 방문자가 데이터를 원하는 형태로 변경할 수 있고, 저장소는 브라우저(즉, 클라이언트가 가지고 있다.)
    - 저장위치: 로컬
    - 보안: 탈취와 변조 가능
    - Lifecycle: 브라우저를 종료해도 파일로 남음
    - 속도: 파일에서 읽기 때문에 상대적으로 빠름

- 세션: 세션도 쿠키를 이용해 데이터를 저장. 하지만 쿠키에는 방문자 고유 ID같은 정보만 저장하고 실제로 모든 데이터는 웹 서버에 저장. 서버에 데이터를 저장하므로 쿠키와는 다르게 저장할 수 있다.
    - 저장위치: 로컬과 서버
    - 보안: ID 값만 가지고 있고 서버에도 저장이 되어 있기 때문에 상대적으로 안전
    - Lifecycle: 브라우저를 종료하면 세션 삭제
    - 속도: 요청마다 서버에서 처리를 해야하기 때문에 비교적 느림
