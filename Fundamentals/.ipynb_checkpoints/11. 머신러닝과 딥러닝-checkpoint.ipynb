### 1. 머신러닝과 딥러닝
#### 1-1. 개념
- 인공지능이란 기계를 지능적으로 만드는 과학으로, 자체 규칙 시스템을 구축한다는 특징이 있다.

- 머신러닝은 인공지능의 하위 개념으로 정확한 결정을 내리기 위해 제공된 데이터를 통하여 스스로 학습할 수 있다. 이 때 빅데이터와 알고리즘을 통해 컴퓨터 자체를 학습시킨다.

- 딥러닝은 인공신경망에서 발전한 형태로, 뇌의 뉴런과 유사한 방식으로 데이터를 학습한다. 이때, 데이터를 스스로 학습한다는 점이 수동으로  데이터를 제공해야하는 머신러닝과 차이를 보인다.

#### 1-2. Representation Learning 데이터 표현 방법
- 1. 분자 형태: 아무런 가공을 거치지 않은 형태 그대로의 표현
- 2. 이미지 형태: 조금 가공된 형태로, 시각적 표현이다. 특정 사이즈를 가지는 2차원 배열로 주로 표현되나, 컬러 이미지라면 RGB 채널을 가지기 때문에 3차원 배열로 표현한다.
- 3. 표 형태: 특징(Feature)을 각 열에 나타내고 열마다 값을 나타낸다. 이때 각 열의 내용을 적합하게 정의하는 과정이 필요하다. (사람의 개입 필요)
- 4. 카테고리: 카테고리는 주로 '예측하고자 하는 값'으로 사용된다. '어떤 카테고리 공간이 있는지', '그 중 어떤 카테고리에 속하는지'를 정하는 과정이 필요하다. (사람의 개입 필요)
- 계층적 관계: 1번에서 4번으로 갈수록 사람의 개입이 강하게 들어간다.

#### 1-3. 딥러닝의 정의
뇌의 신경 구조로부터 영감을 받아 신경망 형태로 설계된 딥러닝의 목표는, '합성된 함수를 학습시켜서 풍부하면서도 유용한 '내재적 표현'을 찾아내는 machine을 구축하는 것'이다.
- 사람의 개입없이 학습된 함수를 사용하여 유용한 내재적 표현을 뽑아내는 것이 목표이다. 

#### 1-4. 데이터의 내재적 표현
데이터의 내재된 표현이란, 그 데이터가 담고 있는 정보의 총체, 혹은 함의를 나타내는 표현을 말한다. 다만, raw한 sensory 데이터는 그 자체로는 중요한 의미를 담지 못하고 있을 수 있다. 마치 이미지를 멀리서 보면 강아지가 보이지만, 각 픽셀 하나하나의 값들은 강아지와 직접적인 연관이 없는 것과 같다.

#### 1-5. 데이터로부터 좋은 표현 추출
데이터로부터 표현을 추출해낸다는 것은 데이터 안에 내재되어 있는, 추상적인 표현을 추출한다는 것을 의미한다. 강아지가 찍힌 이미지에서 좋은 표현을 추출해내려면 픽셀 하나하나의 값으로부터 '강아지'라는 의미를 담고 있는 벡터를 추출해낼 수 있어야 할 것이다.

#### 1-6. 좋은 표현 추출 수행 방법
딥러닝의 본질은 이렇게 데이터로부터 내재된 표현을 추출해내는 것이며, 이는 모델을 Gradient Descent 기반의 학습으로 수행할 수 있다. 물론 더 뛰어난 모델을 얻기 위해서는 모델의 구조나 최적화 기법 등의 세부적인 요인들을 연구하는 것이 필수적이다.

**1-7. 딥러닝의 목표**

데이터의 표현을 학습한다는 것, 특히 사람이 가공하지 않은, 이미지나 자연어 같은 '날 것의' 데이터를 입력하면 알아서 그 안에 내재된 표현을 추출한다는 것

### 2. 딥러닝의 철학
#### 2-1. 행동주의 - 자극, 행동
- 행동주의: 학습된 행동은 자극의 반복으로 만들어진다. 인간의 지능은 살아가면서 받는 자극으로 형성된 후천적인 것이다.
    - 조작적 조건화: 생명체가 외부에서 받는 자극으로 인해 학습되는 과정으로 선택적 보상(강화, 벌)을 이용한다.
    - 스키너 상자(강화 이론 설명): 배고픈 상태의 흰 쥐를 스키너 상자에 넣고 우연히 지렛대를 누르면 먹이가 나오게 한다. 초기에 쥐는 지렛대를 눌러 먹이가 나왔다는 사실을 모르나 우연에 의한 상황이 반복되면 쥐는 지렛대를 누르는 행동을 자주 하게 되며 이를 학습하게 된다. 
    

#### 2-2. 인지주의 - 정보처리

- 행동주의는 인간의 내면을 지나치게 단순화했다는 비판을 받았다. 
- **인간은 동물과 달리 정보를 처리하는 '의식'이 있다.**
- 인간은 뇌 안에서 기억 등의 일련의 정보 처리 과정을 거친 후 행동을 만들어낸다.
- 인지심리학: 지각 과정에서 받아들인 정보가 학습, 기억, 주의 등의 처리과정을 통해 어떻게 심리적으로 가공되고 행동으로 표출되는지를 연구하는 것으로 과학적·기초적 심리학의 한 분야이다. 인간이 지식을 획득하는 방법, 획득한 지식을 구조화하여 축적하는 메커니즘을 주된 연구 대상으로 한다.
- 인지주의는 자극->행동의 관계만 설명하던 행주의에 대해 반발하며 '자극 → (정보처리) → 행동' 의 관점을 제시하였다.
- 인간의 행동이 단순히 자극으로부터 만들어진다는 주장으로 인간의 '의식'을 부정했던 행동주의와 달리, 인지주의는 인간이 자극을 받은 후 내면에서 정보를 처리하고 가공하는 의식이 존재함을 주장하였다. 즉, 인지심리학은 인간의 마음을 일종의 정보 처리 체계로 보고 접근하고, 이에 따라 인간의 뇌가 정보를 처리하는 과정에 대해 관심을 가지며 신경과학이라는 새로운 분야를 만들어냈다.

#### 2-3. 연결주의
- **연결주의: 정보처리는 뇌에 있는 수많은 뉴런의 연결고리를 통해 이루어진다.**
- 인지주의에는 자극으로부터 반응을 하기까지 정보를 처리하는 과정을 수행하는 모형이 있으며, 그 중 하나가 연결주의이다.
- 연결주의는 몇 백억 개의 뉴런이 연결되어 있는 뇌의 형태처럼, 수많은 신호들이 연결된 일련의 과정에 따라 정보가 처리된다고 본다.
- 연결주의의 지능체는 처음에는 '백지' 상태이며, 다수의 사례를 주고 '경험'함으로써 스스로 천천히 '학습'해 간다.
- 학습된 내용은 연결된 뉴런 자체에 저장되어 있으며, 외부에서 자극을 받음에 따라 그 연결 형태가 바뀌면서 학습된 내용이 바뀌어간다.

### 3. 신경망
#### 3-1. 개요
- 딥러닝은 데이터를 입력받으면 일련의 정보 처리 과정을 통해 여러 가지 방법으로 데이터를 표현하다가 원하는 형태로 출력하는 것을 목표로 한다.
- 딥러닝은 연결주의 모형에 따라 정보처리 과정 모형을 뇌와 닮은 신경망으로 채택한다.
- 신경망은 입력을 받은 후 내부에서 일련의 연산 과정을 거쳐 출력을 내는 함수라고 말할 수 있다. 

#### 3-2. 함수
#### 3-2-1. x와 y의 관계 나타내는 도구
- y는 x의 함수이다.
    - y는 x의 변화에 종속적이다. x가 변하면 그 변화하는 정도에 따라 y가 변하게 된다.
    - x가 변하는 정도에 따라 y가 얼마나 변하는지는 x와 y 간 함수의 형태로 결정된다.

#### 3-2-2. (x)를 변환해주는 도구
- 선형 변환

```python
import numpy as np

def linear_transformation(X, A):
    transformed_vector = np.matmul(X, A)
    return transformed_vector

X = np.array([-1,2])
A = np.array([[1,-2],
            [3,0]])
linear_transformation(X, A)
```

### 3-3-3. x의 공간에서 y의 공간으로 매핑 해주는 도구
- 함수의 입력과 출력이 스칼라(Scalar)인지, 벡터(Vector)인지에 따라 구분
- 스칼라는 '크기'만 가지고 '방향'은 가지지 않는 양이다. 주로 단 하나의 숫자로 표현한다.
- 벡터는 '크기'와 '방향'을 두 가지 가지는 양이다. 주로 여러 개의 숫자로 표현되며, 좌표 상에서 길이와 방향을 가지는 화살표로 표현한다.

**1) One-to-One**
- 입력값 x와 출력값 y가 모두 스칼라인 경우. 단 하나의 값을 입력했을 때, 단 하나의 값만 나온다.
$f(x)=wx+b=^y$

**2) Many-to-One**
- Many-to-One은 여러 가지의 정보를 입력받아 하나의 값을 출력하는 경우
- 이 경우 함수는 다음과 같이 벡터인 x를 입력받아 스칼라인 y를 출력하는 형태로 나타낼 수 있습니다. 이때 일반적으로 벡터는 볼드체로 
**x**와 같이 나타낸다.


**회귀문제**
- 입력 데이터 개수에 상관없이, 단 하나의 연속된 실수 스칼라 값을 맞추는 문제
- one-to-one: 단변량 회귀
- many-to-one: 다변량 회귀

**3) Many-to-Many**
- 입력값도 벡터, 출력값도 벡터
- 분류 문제: 여러 개의 카테고리 중 어떤 카테고리에 속하는지를 맞추는 분류 문제는 카테고리 개수로 이루어진 벡터를 출력

**참고**
보통 우리가 MBTI 검사를 할 때에는 여러 개의 질문이 있는 설문지에 답변을 합니다. 그렇다면 위와 같이 성격의 특징 중 '안정을 추구하는 정도', '몽상에 빠지는 빈도' 등과 같은 지표들을 숫자로 체크한 후 모두 모아 벡터 형태로 입력할 수 있겠습니다.

그렇다면 MBTI는 총 4개의 부분에서 E/I, S/N, F/T, P/J를 조합해 네 글자의 결과를 만들어내므로, 네 개의 지표 각각을 0과 1 사이의 숫자로 출력하도록 하여 각각 0에 가까운지, 1에 가까운지에 따라 네 가지의 성향을 결정할 수 있을 것입니다.

여기서 함수는 성격의 여러 가지 특징을 숫자로 표현한 벡터를 입력받은 후 4개의 숫자로 이루어진 벡터를 출력함으로써 MBTI 결과를 만들어내는 것이죠. 실제로 아주 많은 사람들의 설문 결과와 성격 유형에 대한 데이터가 있다면 위와 같은 성격 유형 예측 모델을 만드는 것도 가능할 것입니다.

사실 Many-to-Many가 주로 쓰이는 문제는 바로 '분류' 문제입니다. 위에서 봤던 회귀 문제와 달리, 여러 개의 카테고리 중 어떤 카테고리에 속하는지를 맞추는 분류 문제는 카테고리 개수로 이루어진 벡터를 출력하죠.


#### 2-4. 함수와 모델

### 3. 인공지능

#### 3-1. 선형모델
직선(또는 평면, 초평면 등)의 형태로 정의되는 러닝 모델을 특별히 선형 모델(linear model)이라고 부르며, 이는 가장 단순한 러닝 모델

#### 3-2. 신경망 모델

선형 모델은 가장 단순하고 (선형 모델로 학습이 가능한 데이터 집합에 한해서) 학습 속도가 빠르며 인간이 이해하기 쉬운 대신, 원 데이터 자체가 수많은 픽셀로 이루어져 있는 이미지와 같은 데이터의 경우에는 데이터의 복잡성을 선형 모델에 충분히 반영하는 것이 불가능하다.

그래서, 실제 이미지 인식 분야에서는 선형 모델 대신 인공신경망(artificial neural networks)과 같은 복잡성이 높은 러닝 모델을 사용한다. 이렇게 복잡성이 높은 신경망(neural networks) 구조를 러닝 모델로 사용하는 머신러닝 패러다임을 ‘딥러닝(deep learning)‘이라고 한다.

#### 3-3. 딥러닝
#### 3-3-1. 퍼셉트론
- 심층 신경망의 기본 단위
- 입력 벡터 x=(x1,x2,…,x_d)를 받아들인 뒤 각 성분에 가중치를 곱하는 선형 모델을 거쳐, 그 결과를 모두 합산한 후 활성화함수 σ(⋅)을 적용한 함수로 이루어진 모델이다. 전체 과정을 수행하는 이 선형 모델을 하나의 함수 h(⋅)로 나타낼 수 있으며, 이를 퍼셉트론(perceptron)이라고 부른다.

- 퍼셉트론은 선형 함수와 그 결과를 비선형 함수 활성화 함수까지 거치는 합성 함수

#### 3-3-2. 다층 퍼셉트론
- 퍼셉트론은 세 개의 층으로 이루어진다. 입력 벡터가 자리잡는 층을 입력층(input layer), 최종 출력값이 자리잡는 층을 출력층(output layer), 입력층과 출력층 사이에 위치하는 모든 층을 은닉층(hidden layer)이라고 한다.

- 이 때 퍼셉트론을 기본 빌딩 블록으로 하여, 이런 패턴에 따라 2차원적으로 연결되어 구성되는 인공신경망의 일종을 특별히 다층 퍼셉트론(MLP: multi-layer perceptron)이라고 한다.

#### 3-3-3. 심층 신경망
- 다층 퍼셉트론에서 은닉층의 개수가 많아질수록 인공신경망이 ‘깊어졌다(deep)’고 부르며, 이렇게 충분히 깊어진 인공신경망을 러닝 모델로 사용하는 머신러닝 패러다임을 바로 딥러닝(Deep Learning)이라고 한다.

- 이 때 딥러닝을 위해 사용하는 충분히 깊은 인공신경망을 심층 신경망(DNN: Deep neural network)이라고 한다.

#### 3-3-4. 딥러닝 모델
- 완전 연결 신경망, 컨볼루션 신경망, 순환 신경망

- 컨볼루션 신경망은 주로 이미지 처리에, 순환 신경망은 주로 자연어 처리에 사용된다.

#### 3-3-5. 딥러닝의 장단점
- 장점
    - 요인 표현 학습: 요인 추출의 자동화
        - 심층 신경망의 은닉층의 개수를 늘릴수록 극대화
        - 신속한 커스터마이징 가능
- 단점
    - 많은 데이터 요구량
    - 느린 속도와 높은 요구 사양

#### 3-3-6. 컨볼루션 신경망
컨볼루션 신경망의 낮은 층에 해당할수록, 이미지 상의 각 세부 영역에서의 경계, 명암, 색상 변화 등 저수준(low-level)의 특징들을 요인 표현으로 포착하는 경향을 보인다. 이렇게 위치 별 저수준 요인은, 컨볼루션 신경망의 높은 층으로 올라가면서 서로 가까운 것들끼리 조합되고, 좀 더 넓은 영역에서 고수준(high-level)의 특징들을 요인 표현으로 포착하는 경향을 보인다.

이렇게 낮은 층에서는 점, 선, 면과 같은 간단한 요인들을 추출하고, 점점 높은 층에 다다를수록 복잡한 요인들을 추출하며 최종적으로 이미지를 인식해나간다.

### 4. 딥러닝은 블랙박스인가?

#### 4-1. 블랙박스가 아니다!

신경망은 여러 개의 함수를 엮어 놓은 것이라고 할 수 있고, 그 함수는 본질적으로 회귀 식과 형태가 같다.

즉, 여러 함수의 연산은 간단한 계산으로 이루어져 있고, 그 함수들이 합성된 형태로 수가 어떻게 연산되는지, 어떤 원리로 동작하는지 알기 때문에 블랙박스라고 할 수 없다.

#### 4-2. 블랙박스다!
딥러닝의 구조는 매우 복잡하고 그 연산량 또한 엄청나게 많기 때문에 그 계산 과정을 전체적으로 이해하는 것은 불가능하다. 

학습된 딥러닝 모델이 어떤 결과를 내놓을지 예측할 수 없으며, 모델이 내놓은 결과가 어떤 판단 근거에 따라 판단되었는지 또한 설명할 수 없다.

예를 들어 어느 은행의 인공지능 프로그램이 컴퓨터가 당신의 대출을 거부한 경우, 어떤 근거로 거부 결정을 내렸는지는 실제로 프로그램을 만든 사람조차도 알 수 없다. 이러한 성질로 인해 딥러닝은 블랙박스라고 할 수 있다.


#### 4-3. 머신러닝 모델에 대해 해석할 수 있어야 하는 이유
모델이 예측한 결과가 어떠한 과정으로 도출되었는지 설명할 수 있다면 납득할 만한 과정을 거쳐 도출되었다는 것을 확인함으로써 그 결과를 신뢰할 수 있다. 

상용화 및 대중화를 염두에 두고 머신러닝 기술이 핵심이 되는 제품을 출시하고자 할수록, 제품 사용자가 제품을 신뢰하기 위한 설명의 중요성은 더 높아진다.
